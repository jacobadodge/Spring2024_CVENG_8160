# RNN

```python
input_size = 3
hidden_size = 2
num_layers = 1
output_size = 12
model_arch_rnn = MyRNN(input_size, hidden_size, num_layers, output_size).to(device)
model_arch_rnn.load_state_dict(torch.load('HW02RNN10E.pt'))
model_arch_rnn.eval()
```

```python
results = []
resultsSorted = []
for dataloader in testDataLoaders:
    for idx, data in enumerate(dataloader):
        data = data['inputs'].reshape(-1,36,3)
        data = torch.tensor(np.array([data]*10000),dtype=torch.float32)
        data = data.reshape(-1,36,3)
        result = model_arch_rnn(data.to(device))
        result = (result[0].tolist())
        results.append(result)
# sort the results by detector
step = 87
for i in range(step):
    currentList = []
    for j in range(i, len(results), step):
        currentList.extend(results[j])
    resultsSorted.append(currentList)
```
